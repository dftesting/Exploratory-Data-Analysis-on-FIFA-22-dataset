```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('D:\DATA ANALYST\EXPLORATORY ANALYSIS\players_22.csv',low_memory=False)
df
```

```python
# Check for duplicate data
```

```python
df.duplicated().sum()
```

```python
# drop unused columns
```

```python
df = df.drop(columns=['sofifa_id'])
df = df.drop(columns=['player_url'])
df = df.drop(columns=['long_name'])
df = df.drop(columns=['league_name'])
df = df.drop(columns=['league_level'])
df = df.drop(columns=['club_jersey_number'])
df = df.drop(columns=['club_loaned_from'])
df = df.drop(columns=['club_contract_valid_until'])
df = df.drop(columns=['real_face'])
df = df.drop(columns=['player_tags'])
df = df.drop(columns=['player_face_url'])
df = df.drop(columns=['club_logo_url'])
df = df.drop(columns=['club_flag_url'])
df = df.drop(columns=['nation_logo_url'])
df = df.drop(columns=['nation_flag_url'])
df = df.drop(columns=['nation_team_id'])
```

```python
pd.set_option("display.max_columns", None)
df.columns
```

```python
df
```

```python
klub = df.groupby('nationality_name')['club_name'].nunique()
print(klub)
```

```python
# Top 10 Countries with the Most Number of Clubs
```

```python
top10k = klub.sort_values(ascending=False).head(10)
print (top10k)
```

```python
plt.rc('font', size=8) 
plt.rc('axes', titlesize=8) 
plt.rc('axes', labelsize=8) 
plt.rc('xtick', labelsize=8)
top10k.plot(kind='bar', color='blue')
plt.title('Top 10 Countries with the Most Number of Clubs')
plt.xlabel('Country')
plt.ylabel('Number of Clubs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
club = df.groupby('club_name')['value_eur'].sum()
top10k = club.sort_values(ascending=False).head(10)
print (top10k)
```

```python
# Top 10 Clubs by Total Players Value in FIFA 22
```

```python
top10k.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Club Name')
plt.ylabel('Total Value (EUR)')
plt.title('Top 10 Clubs by Total Players Value in FIFA 22')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Distribution of Player Ages
```

```python
age = df['age'].value_counts().sort_index()
df['age'].plot(kind='hist', bins=20, color='orange',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.title('Distribution of Player Ages')
plt.show()
```

```python
# Does the player's Age correlate with his Wage ?
```

```python
df.plot.scatter(x='age', y='wage_eur', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Wage (EUR)')
plt.title('Age vs. Wage')
plt.show()
```

```python
# Does the player's Overall attribute affect His Wage ?
```

```python
df.plot.scatter(x='overall', y='wage_eur', alpha=0.5)
plt.xlabel('Overall Attribute')
plt.ylabel('Wage (EUR)')
plt.title('Overall Attribute vs Wage')
plt.show()
```

```python
# Top 10 players with the biggest wage
```

```python
wage = df.groupby('short_name')['wage_eur'].sum()
top10w = wage.sort_values(ascending=False).head(10)
top10w.plot(kind='bar', color='green', edgecolor='black')
plt.xlabel('Player Name')
plt.ylabel('Total Wage (EUR)')
plt.title("Top 10 Player's Wage in FIFA 22")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
